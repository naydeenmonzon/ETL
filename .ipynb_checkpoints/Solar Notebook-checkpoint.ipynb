link: https://www.kaggle.com/anikannal/solar-power-generation-data

Description: This data has been gathered at two solar power plants in India over a 34 day period. It has two pairs of files - each pair has one power generation dataset and one sensor readings dataset. The power generation datasets are gathered at the inverter level - each inverter has multiple lines of solar panels attached to it. The sensor data is gathered at a plant level - single array of sensors optimally placed at the plant.

```python
plant1_gen = plant1_gen.rename(columns={'SOURCE_KEY': 'INVERTER_KEY'})
plant1_gen['PLANT_ID'] = plant1_gen['PLANT_ID'].astype("string")
```

```python
plant2_gen = plant2_gen.rename(columns={'SOURCE_KEY': 'INVERTER_KEY'})
plant2_gen['PLANT_ID'] = plant2_gen['PLANT_ID'].astype("string")
```

```python
genCombined = pd.merge(plant1_gen, plant2_gen, how='outer')
genCombined.sort_values(by='DATE_TIME')
```

```python
genCombined.hist(figsize=(16,10))
```

```python
genCombined['DATE_TIME'].max()
```

```python
startTime = genCombined.DATE_TIME.loc[0]
endTime = genCombined.DATE_TIME.loc[1] 
print(endTime-startTime)
```

```python
date = pd.date_range(startTime, endTime,
                     freq='D').to_series()
date.dt.day
```

```python
position = genCombined.columns.get_loc('DATE_TIME')
genCombined['elapsed'] =  genCombined.iloc[1:, position] - genCombined.iat[0, position]
```

```python
genCombined
```

```python
## Summarize the Date by days
genCombined['date'] = genCombined['DATE_TIME'].dt.to_period('D')
```

```python
GenerationDF = genCombined.copy().sort_values(by='TOTAL_YIELD')
GenerationDF
```

```python
GenerationDF.drop(columns=['DATE_TIME', 'elapsed']).hist(figsize=(16,12))
```

```python
GenerationDF['date'] = GenerationDF['date'].astype("object")
```

```python
GenerationDF.groupby('PLANT_ID')['TOTAL_YIELD'].sum().plot(kind='bar', figsize=(16,6),
                                                           title="Plant Total Yeild", fontsize=15, rot=0)
plt.show()
```

```python
GenerationDF.groupby('PLANT_ID').sum()
```

```python
GenerationDF.loc[GenerationDF['DAILY_YIELD']!=0].drop(columns=['TOTAL_YIELD',
        'AC_POWER','DC_POWER']).groupby('INVERTER_KEY').std().sort_values(by='DAILY_YIELD',ascending=False).plot(figsize=(16,8), rot=60, title="Inverter Daily Total Yeild")
plt.show()
```

```python
GenerationDF.loc[GenerationDF['DAILY_YIELD']!=0].drop(columns=['TOTAL_YIELD',
        'AC_POWER','DC_POWER']).groupby('INVERTER_KEY').std().sort_values(by='DAILY_YIELD')
```

```python
date = pd.DataFrame(GenerationDF.groupby('date').agg('mean', 'std'))
date.head()
```
